```python
class Libro:
    def __init__(self, paginas):
        self.paginas = paginas

    def __lt__(self, otro):
        return self.paginas < otro.paginas

    def __eq__(self, otro):
        return self.paginas == otro.paginas

    def __add__(self, otro):
        return self.paginas + otro.paginas


libro1 = Libro(300)
libro2 = Libro(450)
print(libro1 < libro2)  # True
print(libro1 + libro2)
```
